```python
import tensorflow as tf
from tensorflow import keras
from tf_keras_model import get_particle_net
from tensorflow.keras.callbacks import EarlyStopping , ModelCheckpoint
import numpy as np
```

```python
gpu=tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpu[0], True)
```

```python
X_train = np.load('./data/Data_train.npy')
y_train = np.load('./data/Label_train.npy')
X_val = np.load('./data/Data_val.npy')
y_val = np.load('./data/Label_val.npy')
data_train = {'points':X_train[:,:100,:2], 'features': X_train[:,:100,:], 'mask':X_train[:,:100,[2]]}
data_val = {'points':X_val[:,:100,:2], 'features': X_val[:,:100,:], 'mask':X_val[:,:100,[2]]}
```

```python
num_classes = 1
input_shapes = {k:data_train[k].shape[1:] for k in data_train}
model = get_particle_net(num_classes, input_shapes)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.Adam(learning_rate=1e-3),
              metrics=['accuracy'])
model.summary()
```

```python
mcp_save = ModelCheckpoint('ParticleNet_temp.h5', save_best_only=True, monitor='val_loss', mode='min')
```

```python
model.fit(data_train, y_train,
          batch_size=256,
          epochs=20,
          validation_data=(data_val, y_val),
          callbacks = mcp_save,
          shuffle=True)
```
